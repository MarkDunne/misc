```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from scipy.special import gamma
```

```python
def vol(r, n):
    a = np.power(np.pi, n / 2.0)
    b = gamma(n / 2.0 + 1)
    return (r ** n) * a / b 
```

```python
vols = np.array([vol(2, dim) for dim in range(0, 50)])

plt.plot(vols)
plt.xlabel('Dimensions')
plt.ylabel('Volume')
plt.title('Volume of an n-ball with radius 2')
```

```python
max_dim = 50
examples = 10000
dim_avgs = []

points = np.random.normal(size=(examples, max_dim))

for dim in range(1, max_dim+1):
    print 'testing', dim
    dim_pts = points[:, :dim]
    dim_avg = []
    for point in dim_pts:
        avg_dist_from_point = np.sum(np.linalg.norm(dim_pts - point, axis=1) / (examples - 1))
        dim_avg.append(avg_dist_from_point)
    dim_avgs.append(np.mean(dim_avg))
    
print dim_avgs
```

```python
plt.plot(dim_avgs)
plt.xlabel('Dimension')
plt.ylabel('Avg Point Distance')
plt.title('Avg Point Distance per Dimension')
```
