```python
%matplotlib inline

import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from textblob import TextBlob
```

```python
lines = pd.read_csv('databank/movie-dialogs/movie_lines.txt', sep=r" \+\+\+\$\+\+\+ ", names=['lineId','charId','movieId','charName','text'], engine='python')
movies = pd.read_csv('databank/movie-dialogs/movie_titles_metadata.txt', sep=r" \+\+\+\$\+\+\+ ", names=['movieId','title','year','IMDB Rating','vote', 'genres'], engine='python')
conversations = pd.read_csv('databank/movie-dialogs/movie_conversations.txt', sep=r" \+\+\+\$\+\+\+ ", names=['char1','char2','movieId','lineIds'], engine='python')
```

```python
conversations['lineIds'] = conversations['lineIds'].apply(ast.literal_eval)
```

```python
movieId = 'm610'
movies[movies['movieId'] == movieId]
```

```python
def lineIdToInt(line):
    if type(line.lineId) is int:
        return line
    else:
        line['lineId'] = int(line['lineId'][1:])
        return line

movie_lines = lines[lines['movieId'] == movieId]    
# movie_lines = movie_lines.dropna(axis=0, how='any')
# movie_lines = movie_lines.apply(lineIdToInt, axis=1)
# movie_lines = movie_lines.sort('lineId', ascending= True).reset_index(drop=True)
movie_lines[:5]
```

```python
movie_convs = conversations[conversations['movieId'] == movieId]
convs = []
for i, row in movie_convs.iterrows():
    convText = []
    for lineId in row['lineIds']:
        text = movie_lines[movie_lines['lineId'] == lineId].iloc[0].text
        convText.append(text)
    convs.append(' '.join(convText))
    
convs = pd.DataFrame({'text': convs})    
```

```python
def conv_sentiment(conv):
    sentiment = TextBlob(conv['text']).sentiment
    return pd.Series({
        'polarity': sentiment.polarity, 
        'subjectivity': sentiment.subjectivity,
        'convText': conv['text']
    })

analysis = convs.apply(conv_sentiment, axis=1)
analysis = analysis[(analysis['polarity'] != 0) | (analysis['subjectivity'] != 0)]
```

```python
points = analysis.polarity.values * (1 - analysis.subjectivity.values)

points = pd.stats.moments.rolling_mean(points, 10, min_periods=0)

plt.plot(points)
```
